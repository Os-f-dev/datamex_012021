# Proyecto #4

Importar librerias

```python
import pymongo
import pandas as pd
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
```

Conexión con MongoDB

```python
cliente = pymongo.MongoClient()
```

```python
cliente.list_database_names()
```

```python
db = cliente.companies_database
db
```

```python
db.collection_names()
```

```python
colec = db.companies
```

```python
list(colec.find().limit(2))[0].keys()
```

Realizaremos un query y obtendremos empresas fundadas a partir del año 2007 y traeremos los campos( nombre, total de dinero recaudado,oficinas,año de fundado,descripccion,producto,números de empleados,categoria y competencias)

```python
data1=list(colec.find({'$and':[{'founded_year':{'$lte':2007}}]}, 
                         {'name':1, 'total_money_raised':1, 'offices':1,'founded_year':1, 'description':1, 'products':1,'competitions':1,'number_of_employees':1,'category_code':1,'_id':0}))
```

```python
df1=pd.DataFrame(data1)
```

```python
df_2=pd.DataFrame(data1)
```

```python
df1
```

Defino un nuevo Data Frame eliminando de Df1 las columnas products y competitions (para evitar los duplicados )

```python
df_emp=df1.drop(['products','competitions'] , axis=1)
```

```python
lista_data = [json_normalize(i) for i in df_emp['offices']]
```

```python
lista_nombres = list(df_emp['name'])
```

```python
zip_nombre_df = list(zip(lista_nombres,lista_data))
```

```python
offi_df = pd.DataFrame()
```

```python
for i in zip_nombre_df:
    i[1]['name'] = i[0]
```

```python
for i in range(len(lista_data)):
    offi_df = pd.concat([offi_df,zip_nombre_df[i][1]])
```

```python
empresas=pd.merge(df_emp, offi_df, on='name')
```

```python
df_emp.drop(['offices'] , axis=1, inplace=True)
```

```python
empresas.drop(['offices'] , axis=1, inplace=True)
```

```python
empresas.head()
```

```python
plt.figure(figsize=(19, 3))
empresas['country_code'].value_counts()[:900].plot(kind='bar');
```

```python
plt.figure(figsize=(10,10))
empresas['category_code'].value_counts().plot.barh();
```

# Evaluar Nuestra Data

Voy a tomar como referencia USA

```python
empr_def = empresas[empresas['country_code'] == 'USA']
```

```python
plt.figure(figsize=(10,10))
empresas['category_code'].value_counts().plot.barh();
```

Tomaré la Categoria de Analytics

```python
empr_def = empr_def[empr_def['category_code'] == 'analytics']
```

```python
empr_def.head()
```

Ahora teniendo mi data clasificada por el pais y categoria en la que quiero enfocarme , limpiare las columdas products y competitions

aplico nuevamente el paso de offices, ya que ahi tengo un campo importante que es country, no utilizo la data anterior porque ahi habia eliminado products y competitions para evitar duplicados 

```python
lista_da = [json_normalize(i) for i in df_2['offices']]
lista_nomb = list(df_2['name'])
zip_nomd = list(zip(lista_nomb,lista_da))
df_todo = pd.DataFrame()
for i in zip_nomd:
    i[1]['name'] = i[0]

for i in range(len(lista_da)):
    df_todo = pd.concat([df_todo,zip_nomd[i][1]])
    
completa=pd.merge(df_2, df_todo, on='name')
```

Ahora teniendo mi data completa filtro por Country USA y e por Category Analitics

```python
compl_1 = completa[completa['category_code'] == 'analytics'][completa['country_code'] == 'USA']
```

```python
lis_com = [json_normalize(i) for i in compl_1['competitions']]
lis_nom = list(compl_1['name'])
```

```python
zip_nomcom = list(zip(lis_nom,lis_com))
comp_df = pd.DataFrame()
for i in zip_nomcom:
    i[1]['name'] = i[0]
for i in range(len(lis_com)):
    comp_df = pd.concat([comp_df,zip_nomcom[i][1]])
```

```python
comp_df
```

```python
comp_df=comp_df[['name','competitor.name','competitor.permalink']]
```

```python
lis_pro = [json_normalize(i) for i in compl_1['products']]
lis_nm2 = list(compl_1['name'])
zip_npro = list(zip(lis_nm2,lis_pro))
```

```python
pro_df = pd.DataFrame()
for i in zip_npro:
    i[1]['products'] = i[0]
for i in range(len(lis_pro)):
    pro_df = pd.concat([pro_df,zip_npro[i][1]])
    
pro_df = pro_df[['products','name','permalink']]
pro_df.rename(columns={'products':'company','name':'product'}, inplace=True)
```
