In this challenge, we will be using the Pokemon dataset. Before applying statistical methods to this data, let's first examine the data.

To load the data, run the code below.

```python
# Run this code:

pokemon = pd.read_csv('../pokemon.csv')
```

The first thing we would like to do is compare the legendary Pokemon to the regular Pokemon. To do this, we should examine the data further. What is the count of legendary vs. non legendary Pokemons?

```python
# Your code here:

# Conteo si columna legendary = True y otro si Legendary = False
legendary = pokemon[pokemon['Legendary']==True]['Legendary'].count()
non_legendary = pokemon[pokemon['Legendary']==False]['Legendary'].count()
```

```python
print('Legendary:',legendary,'\nNon legendary:',non_legendary)
```

```python
legendary_pokemon = pokemon.loc[pokemon['Legendary']==True]
non_legendary_pokemon = pokemon.loc[pokemon['Legendary']==False]
```

```python
# numero de pokemones legendarios.
legendary_pokemon['Legendary'].count()
```

```python
# número de pokemones no legendarios.
non_legendary_pokemon['Legendary'].count()
```

Compute the mean and standard deviation of the total points for both legendary and non-legendary Pokemon.

```python
# Your code here:

legendary_mean = legendary_pokemon.Total.mean()
legendary_sd = legendary_pokemon.Total.std()

non_legendary_mean = non_legendary_pokemon.Total.mean()
non_legendary_sd = non_legendary_pokemon.Total.std()

print('Legendary')
print(f'mean: {legendary_mean}. standar_deviation: {legendary_sd}')

print('Non Legendary')
print(f'mean: {non_legendary_mean}. standar_deviation: {non_legendary_sd}')
```

The computation of the mean might give us a clue regarding how the statistical test may turn out; However, it certainly does not prove whether there is a significant difference between the two groups.

In the cell below, use the `ttest_ind` function in `scipy.stats` to compare the total points for legendary and non-legendary Pokemon. Since we do not have any information about the population, assume the variances are not equal.

```python
from scipy.stats import ttest_ind
```

```python
# Your code here:

ttest_ind(legendary_pokemon.Total, non_legendary_pokemon.Total, equal_var=False)
```

# Challenge 2 - Matched Pairs Test

In this challenge we will compare dependent samples of data describing our Pokemon. Our goal is to see whether there is a significant difference between each Pokemon's defense and attack scores. Our hypothesis is that the defense and attack scores are equal. In the cell below, import the `ttest_rel` function from `scipy.stats` and compare the two columns to see if there is a statistically significant difference between them.

#### Our null hypothesis is that the defense and attack scores are equal in each Pokemon.

#### Our alternative hypothesis is that the defense and attack scores are different in each Pokemon.

```python
from scipy.stats import ttest_rel
```

```python
ttest_rel(pokemon['Attack'],pokemon['Defense'])
```

Describe the results of the test in the cell below.

```python
# Your conclusions here:

# pvalue < alpha (0.05) => Rechazamos hipótesis nula, pues existe una diferencia 
# estadística entre el ataque y la defena de cada Pokemon.
```

We are also curious about whether therer is a significant difference between the mean of special defense and the mean of special attack. Perform the hypothesis test in the cell below. 

```python
ttest_rel(pokemon['Sp. Atk'],pokemon['Sp. Def'])
```

Describe the results of the test in the cell below.

### Your conclusions here:

        Null hypothesis is not rejected
