## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import numpy as np
```

```python
#importar la dirty data
data = pd.read_csv('../weather-raw.csv')
```

```python
data
```

```python
#obtener toda slas columnas de dias y guardarlas en una lista
dias = []
for col in data.columns:
    if col.startswith("d"):
       dias.append(col)
```

```python
#rellenar valores en null con 0
data1 = data.fillna(0)
#Transformar las columnas de dias en filas
data1 =  pd.melt(data1, id_vars=["id","year","month","element"], value_vars=dias)
data1 = data1.rename(columns={'variable': 'day'})
#Limpiar los valores de los dias
data1 = data1.replace({'day': '^d'}, {'day': ''}, regex=True)
data1 
```

```python
data2 = data1
#transformar las columnas en formato date
data2["date"] = pd.to_datetime(data2[["year","month","day"]],errors='coerce')
#eliminar las columnas individuales
data2= data2.drop(['year', 'month', 'day' ], axis=1)
#eliminar las fechas en null
data2 = data2[data2['date'].notna()]
data2
```

```python
data3= data2.pivot(index='date', columns='element')['value']
data3= data3.rename_axis(columns = None).reset_index()
data3.insert(0, "id", "MX17004")
#dataframe con las columnas reducidas
data3
```

```python
data4 = data3
# vamos a buscar los Tmax y Tmin por mes
# remplazamos los valores 0 porque no interesa al buscar el t.min
data4 = data4.replace(0, np.nan)
#fijamos la fecha como indice para agrupar
data4.set_index('date')
#creamos un periodi para agrupar por mes
per = data4.date.dt.to_period("M")
#guardamos la agrupacion en un dataframe mensual
mensual = data4.groupby(per)
#buscamos los max y min y los guardamos
#la funcion regresa una serie asi que creamos nuevos dataframe
tmax = pd.DataFrame(mensual['tmax'].max())
tmin= pd.DataFrame(mensual['tmin'].min())
#hacemos un merge de ambos dataframe
result = tmax.join(tmin, how="inner")
result = result.rename_axis(columns = None).reset_index()
result
```
