```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
df=pd.read_csv('attacks.csv', encoding='latin-1')
```

```python
#Se lee el archivo y se analiza el tipo de dato de cada columna
```

```python
df.info()
```

```python
df.drop([ 'Unnamed: 22', 'Unnamed: 23', 'href formula' , 'pdf', 'Case Number.1', 'Case Number.2'] , axis = 1, inplace = True)
```

```python
#eliminacion de columnas que se consideran no importantes y eliminacion de todas las
#filas con valores nulos y con 0 y xx
```

```python
df=df.dropna(how='all')
```

```python
indexNames = df[ (df['Case Number'] == '0') ].index
df.drop(indexNames , inplace=True)
indexNames = df[ (df['Case Number'] == 'xx') ].index
df.drop(indexNames , inplace=True)
```

```python
df.head()
```

```python
df['Case Number']=df['Case Number'].astype('str')
```

```python
#Cambio el tipo de dato de Case number para poder tomar solo el año.
```

```python
df['Date'] = df['Case Number'].apply(lambda x: "-".join(re.findall('(^\d+)',x)))
```

```python
#Hago copia de mi DataFrame
df2=df.copy()
df2
```

```python
df2.drop(['Year'] , axis = 1, inplace = True)
```

```python
#Elimino la columna Year y luego asigno original order a Case number y luego la elimino
```

```python
df2['Case Number'] = df2['original order']
```

```python
df2.drop(['original order'] , axis = 1, inplace = True)
```

```python
df2
```

```python
indexX = df2[ (df2['Date'] < '2') ].index
df2.drop(indexX , inplace=True)
```

```python
#Elimino de mi columna Date los menores a 2 para eliminar data vieja menores a el año 2000
#Luego convierto Case number y Date a entero.
```

```python
df2['Case Number']=df2['Case Number'].astype(int)
```

```python
df2['Date']=df2['Date'].astype(int)
```

```python
df2.info()
```

```python
df2.drop(['Age', 'Time'] , axis = 1, inplace = True)
```

```python
df2
```

```python
column_order = ['Case Number','Date','Name','Country',
                'Area','Location','Sex ',
                'Fatal (Y/N)','Activity','Species ',
                'Investigator or Source','href']
 
df2 = df2[column_order]
df2.head()
```

```python
indexNames = df2[ (df2['Name'] == 'male') ].index
df2.drop(indexNames , inplace=True)
```
