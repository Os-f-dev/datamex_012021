# HTML 

```python
from bs4 import BeautifulSoup as bs
from lxml import html
```

```python
url1 = 'https://es.wikipedia.org/wiki/Age_of_Empires_II:_The_Age_of_Kings'
html = requests.get(url1).content
soup = bs(html, 'lxml')
soup
```

```python
table = soup.findAll('table',{'class':'wikitable'})
table
```

```python
tablas = table[2]
tablas
```

```python
cln = tablas.findAll('th')
cln
```

```python
cln_txt = [i.text.replace('\n','') for i in cln]
cln_txt
```

```python
rows = tablas.findAll('tr')
rows
```

```python
for i in rows:
    print(i.text.strip().split('\n'))
```

```python
rows_clean = [i.text.strip().split('\n') for i in rows]
rows_clean
```

```python
rows2 = rows_clean.copy()
rows2
```

```python
rows2.pop(0)
```

```python
tabla_final = []

for i in rows2:
    temp = []
    for j in i:
        if j != '':
            temp.append(j)
        tabla_final.append(temp)
tabla_final
```

```python
df_final = pd.DataFrame(tabla_final, columns = cln_txt)
df_final
```

# CONCAT

```python
final = pd.concat([df, df_final], axis = 1)
final
```
